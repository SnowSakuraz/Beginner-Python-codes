```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(132)
comprehensive = list()

for i in range(5):    
    random_walk = [0]
    for i in range(100):
        step = random_walk[-1]
        num = np.random.randint(1, 7)
        random_walk.append(step + num)
    comprehensive.append(random_walk)

array = np.transpose(np.array(comprehensive))
plt.style.use("seaborn")
plt.plot(array)
plt.xlabel("numbers of dice-rolling")
plt.ylabel("total points")
plt.show()
```
