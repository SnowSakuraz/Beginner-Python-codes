```python
# Random walk 
import numpy as np
heads = [0]
for i in range(10):
    coin = np.random.randint(0,2)
    heads.append(heads[i] + coin)   #注意append是增加
print(heads)

# Random walk 2 
random = [0]
for i in range(10):
    step = random[-1]    #这一步是为了让数字变为incremental
    num = np.random.randint()
    random.append(step + num)
    
    
```

```python
import numpy as np, matplotlib.pyplot as plt

all_walks = []
for i in range(15):   # run 15 times
    random_walk = [0]
    for i in range(100):    # roll the dice 100 times
        step = random_walk[-1]   
        
        # 掷骰子 与 游戏规则
        dice = np.random.randint(1,7)
        if dice <=2:
            step = max(step - 1, 0)
        elif dice <=5:
            step += 1
        else:
            step = step + np.random.randint(1,7)
            
        random_walk.append(step)
    all_walks.append(random_walk)
    
np_ar = np.array(all_walks)
plt.plot(np_ar)
plt.show()
```
